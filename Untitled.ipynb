```python
train = []
test = []
for root, subdirs, files in os.walk(path+'/images/original_videos/'):
    if len(files) == 0: continue
    idx = root.index('original_videos/')
    video_pth = root[idx:]+'.mp4'
    if video_pth in meta:
        info = meta[video_pth]
        files = [{'pth': '/'.join([root, f]), 'label': info['label']} for f in files]
        if info['set'] == 'train':
            train.extend(files)
        else:
            test.extend(files)
    else:
        print(video_pth)
        1/0
len(train), len(test)
#     print(root, files)
#     output.extend(files)
```
